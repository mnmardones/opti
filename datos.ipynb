```python
import numpy.random as nr
import numpy as np
import json
```

```python
import pandas
```

```python
alfa, gamma = (round(x,2) for x in nr.random(2))
tractor = nr.choice([0,1], 40, p=[0.7, 0.3])
afinidad = nr.choice([0,1], [40, 40])
afinidad = np.tril(afinidad) + np.tril(afinidad, -1).T
experiencia = [round(min(nr.exponential(0.2), 1),2) for i in range(40)]
```

```python
print(alfa, gamma)
pandas.DataFrame(tractor)
```

```python
pandas.DataFrame(afinidad)
```

```python
pandas.DataFrame(experiencia)
```

```python
datos = {'alfa': alfa, 'gamma': gamma, 'tractor': tractor.tolist(), 'afinidad': afinidad.tolist(), 'experiencia': experiencia}
with open('datos.json','w') as f:
    json.dump(datos, f)
```
